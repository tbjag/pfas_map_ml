```python
import rasterio
import matplotlib.pyplot as plt
from rasterio.plot import show
import numpy as np

from matplotlib.colors import Normalize
```

```python
# Open the raster file
with rasterio.open('test_out.tif') as src:
    # Read the data
    raster_data = src.read(1)  # Read the first band

    # Mask NoData values (assuming the NoData value is already set in the raster)
    raster_data = np.ma.masked_equal(raster_data, src.nodata)

    # Plot the raster data with a colorbar and improved colormap
    fig, ax = plt.subplots(figsize=(10, 8))

    # Normalize the data (optional) if you want to scale the intensity
    norm = Normalize(vmin=raster_data.min(), vmax=raster_data.max())

    # Use a colormap like 'viridis', 'plasma', or 'inferno' to better visualize intensity variations
    cax = ax.imshow(raster_data, cmap='viridis', norm=norm)

    # Add a colorbar to show the AQI scale
    fig.colorbar(cax, ax=ax, label='AQI Intensity')

    # Optionally overlay raster bounds
    # show(src, ax=ax, cmap='gray', alpha=0.3)  # Using alpha to overlay with transparency

# Show the plot
plt.show()
```
