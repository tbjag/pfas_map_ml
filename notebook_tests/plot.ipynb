```python
import rasterio
import matplotlib.pyplot as plt
from rasterio.plot import show
import numpy as np

from matplotlib.colors import Normalize
```

```python
# Open the raster file
with rasterio.open('/home/ciom/workspace/pfas_map_ml/data/output/temp.tiff') as src:
    # Read the data
    raster_data = src.read(1)  # Read the first band

    # Mask NoData values (assuming the NoData value is already set in the raster)
    raster_data = np.ma.masked_equal(raster_data, src.nodata)

    # Print the first 10 pixels
    print("First 10 pixels (flattened):")
    print(raster_data.flatten()[:10])

    # Plot the raster data with a colorbar and improved colormap
    fig, ax = plt.subplots(figsize=(10, 8))

    # Normalize the data (optional) if you want to scale the intensity
    norm = Normalize(vmin=raster_data.min(), vmax=raster_data.max())

    # Use a colormap like 'viridis', 'plasma', or 'inferno' to better visualize intensity variations
    cax = ax.imshow(raster_data, cmap='inferno')

    # Get min and max values of the raster data
    min_value = np.min(raster_data)
    max_value = np.max(raster_data)
    u = np.unique(raster_data)

    print(f"\nRange of values: min = {min_value}, max = {max_value}")
    print(f"Unique values: {u}")

# Show the plot
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Step 1: Load the shapefile
shapefile_path = '../HUC8_CA_PFAS_GTruth.shp'
gdf = gpd.read_file(shapefile_path)

# Step 2: Create a figure with specified size
plt.figure(figsize=(10, 8))

# Step 3: Plot the shapefile
gdf.plot()

plt.title('Shapefile Plot')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
